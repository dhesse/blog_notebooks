```python
import pandas
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
data = pandas.concat((pandas.read_csv(f"bysykkel-0{i}.csv.gz", parse_dates=[0,1]) for i in [4,5]))
```

```python
(data[data.duration < 30*60].duration / 60).plot.hist(bins=30)
```

```python
data.groupby('started_at').size()
```

```python
data.groupby('started_at').size().resample('1 h').sum().plot.line()
```

```python
data.groupby('started_at').size().resample('1 h').sum()['2022-05'].plot.line()
```

```python
data.set_index('started_at').resample('1h').duration.mean()['2022-05'].plot.line()
```

```python
data.set_index('started_at').resample('1h').duration.median()['2022-05'].plot.line()
```

```python
data.set_index('started_at').groupby([pandas.Grouper(freq='1h'), pandas.Grouper(freq='1d')]).size()
```

```python
by_h = data.set_index('started_at').groupby([lambda x: x.hour, lambda x: x.date]).size().reset_index()
by_h.columns = ['hour', 'day', 'count']
by_h = by_h.pivot(values='count', index='day', columns='hour').fillna(0)
by_h.T.plot.line(alpha=0.5, legend=False)
```
